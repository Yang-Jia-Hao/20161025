```python
import numpy as np
K = 9100
Interval = 500 #Range
Premium_Call = 179
Premium_Put = 185
ST = np.arange(K - Interval, K + Interval) #S0~ST期 X軸
```

```python
#payoff 計算 向量化運算
Payoff_LongCall = np.maximum(ST - K, 0) - Premium_Call #Long Call 買買權
Payoff_ShortCall = - Payoff_LongCall                   #Short Call 賣買權  
```

```python
import matplotlib.pyplot as plt #簡化叫plt
plt.plot(ST, Payoff_LongCall)
plt.plot(ST, Payoff_ShortCall)
plt.plot(ST, Payoff_LongCall, label=u'$+C$' ,color="red" ,linewidth=2) # 加上label參數添加圖例
plt.plot(ST, Payoff_ShortCall, label=u'$-C$',color="blue" ,linewidth=8 )
plt.xlabel("123")
plt.ylabel('456')
plt.grid()   #網格
plt.legend() #加這個我的圖例才出來
```

```python
Payoff_LongPut = np.maximum(K - ST, 0) - Premium_Put 
Payoff_ShortPut = - Payoff_LongPut
```

```python
plt.figure
plt.plot(ST, Payoff_LongPut)
plt.plot(ST, Payoff_ShortPut)
plt.plot(ST, Payoff_LongPut, label=u'$+P$' ,color="red",linewidth=2) # 加上label參數添加圖例
plt.plot(ST, Payoff_ShortPut, label=u'$-P$',color="blue",linewidth=8)
plt.legend()
```
